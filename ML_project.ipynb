Now we handle null values.

Since thare are no missing values in the traget variable for both test and train set, we do not have to handle null values for target variables. Here we first handle missing values for numerical features.

```python
# For numerical features, we can use the median to impute missing values
X_train[numerical_features] = X_train[numerical_features].fillna(X_train[numerical_features].median())
X_test[numerical_features] = X_test[numerical_features].fillna(X_test[numerical_features].median())
```

Then handle missing values for categorical variables.

```python
# For categorical features, impute missing values with the mode
X_train[categorical_features] = X_train[categorical_features].fillna(X_train[categorical_features].mode().iloc[0])
X_test[categorical_features] = X_test[categorical_features].fillna(X_test[categorical_features].mode().iloc[0])
```

Now we handle the outliers.

For that we use capping outliers instead of removing the outliers, since we lack of data.

```python
# For each numerical feature, cap the outliers to the 97th and 3rd percentiles
for feature in numerical_features:
    upper_limit = X_train[feature].quantile(0.97)
    lower_limit = X_train[feature].quantile(0.03)
    X_train[feature] = X_train[feature].clip(lower=lower_limit, upper=upper_limit)
    X_test[feature] = X_test[feature].clip(lower=lower_limit, upper=upper_limit)
```

### Handle Categorical Values

Since some labels in the train dataset not present in the test dataset we don't use the Label Encoding method. Therefore, we use one hot encoding method to handle categorical data.

```python
import pandas as pd

# Apply One-Hot Encoding to categorical features in training and testing sets
X_train_encoded = pd.get_dummies(X_train, columns=categorical_features, drop_first=True)
X_test_encoded = pd.get_dummies(X_test, columns=categorical_features, drop_first=True)

# Ensure the columns in X_train_encoded and X_test_encoded match
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1)

# Fill missing values in the test set with 0 after alignment
X_test_encoded = X_test_encoded.fillna(0)

# Now X_train_encoded and X_test_encoded have the same columns, and categorical variables are encoded

# For target variable
if y_train.nunique() > 2:
    y_train_encoded = pd.get_dummies(y_train, drop_first=True)
    y_test_encoded = pd.get_dummies(y_test, drop_first=True)
else:
    # If y_train is binary, you can directly use the numeric values (0 or 1)
    y_train_encoded = y_train
    y_test_encoded = y_test
```

Then we handle birnary variables for target variables.

```python
# Map the 'Yes' and 'No' values to 1 and 0 for the target variable
y_train_encoded = y_train.map({'Yes': 1, 'No': 0})
y_test_encoded = y_test.map({'Yes': 1, 'No': 0})
```

```python
# Display the first 5 rows of the dataset after One-Hot Encoding as a table
import pandas as pd

# Show the encoded training set as a table
print("Training Set After One-Hot Encoding:")
display(X_train_encoded.head())

# Show the encoded testing set as a table
print("\nTesting Set After One-Hot Encoding:")
display(X_test_encoded.head())

print("Columns in the dataset:")
print(X_train_encoded.columns)



```

### Scale the Numerical Features

To standardize (scale) the numerical features using the StandardScaler

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale the training dataset (compute mean and std from training data)
X_train_encoded[numerical_features] = scaler.fit_transform(X_train_encoded[numerical_features])

# Scale the test dataset (use the mean and std from training data to scale test data)
X_test_encoded[numerical_features] = scaler.transform(X_test_encoded[numerical_features])
```

### Handling Imbalanced Dataset

Handle the imbalanced dataset using SMOTE.

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Initialize SMOTE for oversampling the minority class
smote = SMOTE(random_state=42)

# Apply SMOTE to the training data
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_encoded, y_train_encoded)

# Check the class distribution before and after resampling
print(f"Original class distribution: {Counter(y_train_encoded)}")
print(f"Resampled class distribution: {Counter(y_train_resampled)}")
```

# **Machine Learning Model Development & Validation**

## **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import numpy as np


# Initialize the Logistic Regression model with more iterations
logreg = LogisticRegression(random_state=42, max_iter=1000)

```

We used 5-Fold Cross-Validation in logistic regression to make sure the model works well on different parts of the data and not just the specific training data. It splits the training data into 5 parts, trains on 4 parts, and tests on the remaining part, repeating this 5 times. This helps check how reliable and consistent the model is.

```python
# Perform 5-fold cross-validation on the training data
cv_scores = cross_val_score(logreg, X_train_encoded, y_train_encoded, cv=5, scoring='accuracy')

# Print cross-validation scores
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Average Cross-Validation Accuracy: {np.mean(cv_scores):.4f}")
```

```python
# Fit the Logistic Regression model on the training data
logreg.fit(X_train_encoded, y_train_encoded)

# Logistic Regression: Predict on the test set
logreg_predictions = logreg.predict(X_test_encoded)

# Evaluate Logistic Regression model on test set
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Logistic Regression Model Evaluation on Test Set:")
print(f"Accuracy: {accuracy_score(y_test_encoded, logreg_predictions):.4f}")
```

```python
print("Classification Report:")
print(classification_report(y_test_encoded, logreg_predictions))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test_encoded, logreg_predictions))
```

## Random Forrest

We used Hold-Out Validation in the random forest model to keep things simple. We trained the model on one part of the data (training set) and tested it on a completely separate part (test set) to see how well it performs. This helps us check if the model works well on new, unseen data.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define the model
rf = RandomForestClassifier(random_state=42)

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [10, 20, None],       # Maximum depth of each tree
    'min_samples_split': [2, 5],       # Minimum samples required to split a node
    'min_samples_leaf': [1, 2],        # Minimum samples required to be a leaf node
}

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the grid search to the resampled training data
grid_search.fit(X_train_resampled, y_train_resampled)

# Print the best parameters found by Grid Search
print(f"Best Parameters: {grid_search.best_params_}")

# Get the best model from Grid Search
best_rf = grid_search.best_estimator_

# Evaluate on Training Set
train_accuracy = best_rf.score(X_train_resampled, y_train_resampled)
print(f"Training Accuracy: {train_accuracy:.4f}")

# Evaluate on Test Set
test_accuracy = best_rf.score(X_test_encoded, y_test_encoded)
print(f"Test Accuracy: {test_accuracy:.4f}")

# Predictions
y_train_preds = best_rf.predict(X_train_resampled)
y_test_preds = best_rf.predict(X_test_encoded)
```

```python
# Classification Report and Confusion Matrix for Training Set
print("\nRandom Forest Model Evaluation on Training Set:")
print("Classification Report:")
print(classification_report(y_train_resampled, y_train_preds))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_train_resampled, y_train_preds))
```

```python
# Classification Report and Confusion Matrix for Test Set
print("\nRandom Forest Model Evaluation on Test Set:")
print("Classification Report:")
print(classification_report(y_test_encoded, y_test_preds))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test_encoded, y_test_preds))
```
